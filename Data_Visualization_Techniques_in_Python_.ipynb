```python
sal = [15,22,24,25,30,23,34,43,19,20,52,21]
sal.sort()
sal
```

OUTLIERS

```python
import pandas as pd
ser = pd.Series(sal)
```

```python
q1 =ser.quantile(0.25)
print(q1)
```

```python
q2 = ser.quantile(0.50)
print(q2)
```

```python
q3 = ser.quantile(0.75)
print(q3)
```

```python
# IQR inner quatile range
IQR = q3-q1
print(IQR)
```

```python
# outlier detection formula
Higher_outlier = (IQR * 1.5) + q3

Lower_outlier = (IQR *1.5) - q1


print("Higher_outlier =",Higher_outlier)
print("Lower_outlier =",Lower_outlier)
```

SCATTER PLOT

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pandas.plotting import scatter_matrix
```

```python
iris = sns.load_dataset('iris')
df = iris.drop(['species'],axis=1)

df.head(3)
```

```python
# Kde = Kernel density estimation

scatter_matrix(df,figsize=(10,8),diagonal="kde", color ='r')
plt.show()
```

```python
# Without Kde plot is hist plot
scatter_matrix(df,figsize=(10,8), color ='b')
plt.show()
```
