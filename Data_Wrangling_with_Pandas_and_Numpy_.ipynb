```python
# Create dataframe with missing values

raw_data = {
    "first_name" : ["Jason",np.nan, "Tina", "Jake", "Amy", "Madhu"],
    "Last_name" :["Millar",np.nan,"Ali","Milner","Cooze","Patil"],
    "age" : [42,np.nan,36,24,73,25,],
    "sex" : ["M",np.nan,"F","M","F","F"],
    "unit-1": [4,np.nan,np.nan,2,3,4],
    "unit-2": [3,np.nan,np.nan,4,4,4]}

df = pd.DataFrame(raw_data, columns =['first_name','Last_name','age','sex','unit-1','unit-2'])

df
```

```python
df_no_missing = df.dropna()
df_no_missing
```

```python
df  # not saved permantly not (implace = true) # Axis = 0 columns wise
```

```python
# Rows that has all NAN's will be removed
df_cleaned = df.dropna(how="all")
df_cleaned
```

```python
# Create a new column full of missing values

df["Final_Score"] =np.nan

df
```

```python
# Drop column if they only contain missing values

df.dropna(axis=1, how="all") # axis=1 means drop columns cotains all where nan value

```

```python
df.fillna(0)  # fill nan value with 0 just see but we can not fill 0 nonsense
```

```python
df
```

```python
# Fill in missing in unit-1 with the mean value of unit-1

# inplace = True means that the changes are saved to the df right away

df["unit-1"].fillna(df["unit-1"].mean(),inplace=True)
df
```

```python
# fill in missing in postTestScore with each sex's mean value of postTestScore

df["unit-2"].fillna(df.groupby("sex")["unit-2"].transform("mean"),inplace=True)
df
```

```python
# Select some raws but ignore the missing data point
df[df['age'].notnull() & df['sex'].notnull()]
```

```python
df
```

```python
# back-fill
df.fillna(method='bfill')

```

```python
# forword_fill
df.fillna(method="ffill",inplace=True)
df
```

```python

# total clean up data
df['Final_Score'].fillna(df['unit-1'] + df['unit-2'],inplace=True)
df
```

```python
import pandas as pd
import numpy as np
titanic_train = pd.read_csv("https://raw.githubusercontent.com/rolandmueller/titanic/main/titanic3.csv")
titanic_train.head()
```

```python
# How to check unique values of any colunm

titanic_train["survived"].unique() # similary you can check any column unique valves
```

```python
# How to the rem=name categories using pandas method
new_survived = pd.Categorical(titanic_train['survived'])
new_survived = new_survived.rename_categories(["Died","Survived"])
new_survived.describe()
```

```python
titanic_train["pclass"].unique()
```

```python
new_pclass = pd.Categorical(titanic_train["pclass"],ordered=True)  # converting 1,2,3 into class1,class2,class3


new_pclass = new_pclass.rename_categories(["Class1","Class2","Class3"])

new_pclass.describe()
```

```python
titanic_train["pclass"] = new_pclass # Reassing new name to pclass

titanic_train.head()
```

```python
titanic_train["cabin"].unique() # check unique cabin
```

```python
char_cabin = titanic_train["cabin"].astype(str) # convert data to str

# List comprehension (Creating list using another set of string,list etc)

new_cabin = np.array([cabin [0] for cabin in char_cabin]) # Take First letter

new_cabin = pd.Categorical(new_cabin)

new_cabin.describe()
```

```python
titanic_train["cabin"] = new_cabin
```
